# Age-Based Discount Checker

This Python program checks whether a person is eligible for a 10% discount based on age.

## Criteria
- Age between 22 and 30 → Eligible
- Otherwise → Not Eligible

## Concepts Used
- Lists
- for loop
- zip()
- Conditional statements

## Output Example
Navin, you are eligible for a 10% discount

```python
Name=["Navin","praveen","sanam","gokul","mani"]
Age=[22,27,32,15,28]

for n, b in zip(Name,Age):
  if b>= 22 and  b<= 30:
    print(n,"your are elegible for 10% discount")
  else:
    print(n,"your not elegible")
```
